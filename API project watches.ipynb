```python
import requests

url = "https://real-time-amazon-data.p.rapidapi.com/search"

querystring = {"query":"Watches","page":"1","country":"IN","category_id":"aps"}

headers = {
	"X-RapidAPI-Key": "cddeb5394fmsh40d77ed851109fbp16d60bjsnd50925bd29af",
	"X-RapidAPI-Host": "real-time-amazon-data.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
import numpy as np
import pandas as pd
```

```python
a=pd.json_normalize(response.json()["data"]["products"])
```

```python
a
```

```python
product_id=[i for i in a["asin"]]
```

```python
product_id
```

```python
l=[]
for i in product_id:
    url = "https://real-time-amazon-data.p.rapidapi.com/product-reviews"
    querystring = {"asin":i,"query":"Watches","page":"1","country":"IN","category_id":"aps"}

    headers = {
        "X-RapidAPI-Key": "cddeb5394fmsh40d77ed851109fbp16d60bjsnd50925bd29af",
        "X-RapidAPI-Host": "real-time-amazon-data.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)

    l.append(response.json())
```

```python
l
```

```python
reviews=pd.DataFrame()
for i in range(len(l)):
    df=pd.json_normalize(l[i]["data"]["reviews"])
    reviews=pd.concat([reviews,df])
```

```python
reviews=reviews[["review_comment","review_star_rating"]]
```

```python
reviews=reviews.reset_index(drop=True)
```

```python
reviews
```

```python
reviews.to_csv("watches.csv",index=None)
```
